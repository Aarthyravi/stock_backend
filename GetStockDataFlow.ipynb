```python
def get10dayavg(conn,symbolid):
  
    cur = conn.cursor()
    cur.execute("SELECT * FROM (SELECT * FROM stockprices order by id DESC limit 10) " 
                "where stockdetails_id = ? order by id",(symbolid, ))
    
    c = len(cur.fetchall()) 
      
    if c < 10:
        return -1 
            
    elif c == 10:
        cur = conn.cursor()
        c = cur.execute("SELECT sum(volume)/count(*) FROM stockprices "
                        "where stockdetails_id = ?",(symbolid,))
        for row in c:
            return row[0]
        
def get50dayavg(conn,symbolid):
  
    cur = conn.cursor()
    cur.execute("SELECT * FROM (SELECT * FROM stockprices order by id DESC limit 50) "
                "where stockdetails_id = ? order by id",(symbolid, ))
    
    c = len(cur.fetchall()) 
      
    if c < 50:
        return -1 
            
    elif c == 50:
        cur = conn.cursor()
        c = cur.execute("SELECT sum(volume)/count(*) FROM stockprices "
                        "where stockdetails_id = ?",(symbolid,))
        for row in c:
            return row[0]

def get100dayavg(conn,symbolid):
  
    cur = conn.cursor()
    cur.execute("SELECT * FROM (SELECT * FROM stockprices order by id DESC limit 100) "
                "where stockdetails_id = ? order by id",(symbolid, ))
    
    c = len(cur.fetchall()) 
      
    if c < 100:
        return -1 
            
    elif c == 100:
        cur = conn.cursor()
        c = cur.execute("SELECT sum(volume)/count(*) FROM stockprices "
                        "where stockdetails_id = ?",(symbolid,))
        for row in c:
            return row[0]        

def get365dayavg(conn,symbolid):
  
    cur = conn.cursor()
    cur.execute("SELECT * FROM (SELECT * FROM stockprices order by id DESC limit 365) "
                "where stockdetails_id = ? order by id",(symbolid, ))
    
    c = len(cur.fetchall()) 
      
    if c < 365:
        return -1 
            
    elif c == 365:
        cur = conn.cursor()
        c = cur.execute("SELECT sum(volume)/count(*) FROM stockprices "
                        "where stockdetails_id = ?",(symbolid,))
        for row in c:
            return row[0]

def getminvolume(conn,symbolid):
    
    cur = conn.cursor()
    minvolume = cur.execute("SELECT min(volume) FROM stockprices "
                            "where stockdetails_id =?",(symbolid, ))
                
    for row in minvolume:
        return row[0]

def getmaxvolume(conn,symbolid):
    
    cur = conn.cursor()
    maxvolume = cur.execute("SELECT max(volume) FROM stockprices "
                            "where stockdetails_id =?",(symbolid, ))
                
    for row in maxvolume:
        return row[0] 

def get50dayavgprice(conn, symbolid):
    cur = conn.cursor()
    cur.execute("SELECT * FROM (SELECT * FROM stockprices order by id DESC limit 50) "
                "where stockdetails_id = ? order by id",(symbolid, ))
    
    c = len(cur.fetchall()) 
      
    if c < 50:
        return -1 
            
    elif c == 50:
        cur = conn.cursor()
        c = cur.execute("SELECT sum(closevalue)/count(*) FROM stockprices "
                        "where stockdetails_id = ?",(symbolid,))
        for row in c:
            return row[0]
    
def get100dayavgprice(conn, symbolid):
    cur = conn.cursor()
    cur.execute("SELECT * FROM (SELECT * FROM stockprices order by id DESC limit 100) "
                "where stockdetails_id = ? order by id",(symbolid, ))
    
    c = len(cur.fetchall()) 
      
    if c < 100:
        return -1 
            
    elif c == 100:
        cur = conn.cursor()
        c = cur.execute("SELECT sum(closevalue)/count(*) FROM stockprices "
                        "where stockdetails_id = ?",(symbolid,))
        for row in c:
            return row[0]    
    
def get_symbol_data_today(now,sym):
    
    #start=%s/%s/%s" % (i.day, i.month, i.year)
    prev = datetime.datetime.now() - timedelta(days=3)
    start=datetime.datetime(prev.year,prev.month,prev.day)
    end=datetime.datetime(now.year,now.month,now.day)
    try:
        f = web.DataReader(sym, 'yahoo', start,end)
    except Error as e:
        print(e)    
    
    #f=f.set_index(['Date'])
    #r=web.get_quote_yahoo('AAPL')
    return f.tail(1)

def update_data_to_db(conn,sym):
    now = datetime.datetime.now()
    data = get_symbol_data_today(now,sym)
    #print(data.keys())
    t=data['Open'].values[0]
    t1=data['High'].values[0]
    t2=data['Low'].values[0]
    t3=data['Close'].values[0]
    t4=data['Adj Close'].values[0]
    t5=data['Volume'].values[0]
                            
    symbolid = get_symbol_id(conn, sym)
    
    stockpricesinfo = (datetime.date.today(),symbolid,t,t1,t2,t3,t4,
                       int.from_bytes(t5, byteorder='little',))
    insert_stockprices(conn, stockpricesinfo) 
    
    avg10volume = get10dayavg(conn,symbolid)
    avg50volume = get50dayavg(conn,symbolid)
    avg100volume = get100dayavg(conn,symbolid)
    avg365volume = get365dayavg(conn,symbolid)
    minvolume = getminvolume(conn,symbolid)
    maxvolume = getmaxvolume(conn,symbolid)
    day50avgprice = get50dayavgprice(conn,symbolid)
    day100avgprice = get100dayavgprice(conn,symbolid)
    stockpricesinfo1 = (datetime.date.today(),symbolid,t,t1,t2,t3,t4,
                        int.from_bytes(t5, byteorder='little'),avg10volume,
                        avg50volume,avg100volume,avg365volume,minvolume,maxvolume,
                        day50avgprice,day100avgprice,datetime.date.today(),symbolid)
    id = update_stockprices(conn, stockpricesinfo1)

    
def update_currentvalue(conn, sym):
    now = datetime.datetime.now()
    data = get_symbol_data_today(now,sym)
    #print(data.keys())
    t3=data['Close'].values[0]
    return t3
    
def get_symbol_id(conn,sym):
    
    cur = conn.cursor()
    cur.execute("SELECT id FROM stockdetails WHERE companysymbol=?",(sym,))
    row=cur.fetchone()
    return row[0] 

def get_symbol(conn,id):
    
    cur = conn.cursor()
    cur.execute("SELECT companysymbol FROM stockdetails WHERE id=?",(id,))
    row=cur.fetchone()
    return row[0] 

   
def insert_stockprices(conn,stockpricesinfo):    
    """
    Insert a new stockprices into the stockprices table
    :param conn:
    :param stockprices:
    
    """
    cur = conn.cursor()
    count = cur.execute("SELECT count(*) FROM stockprices WHERE date = ? and stockdetails_id =?",
                        (datetime.date.today(),stockpricesinfo[1]))
    
    for row in count:
        
        if (row[0] == 0):
            sql = ''' INSERT INTO stockprices(date,stockdetails_id,openvalue,highvalue,lowvalue,
                    closevalue,adjclose,volume) VALUES(?,?,?,?,?,?,?,?) ''' 
            cur.execute(sql, stockpricesinfo)
    
    
def update_stockprices(conn,stockpricesinfo1):    
    """
    Update a old stockprices into the stockprices table
    :param conn:
    :param stockprices:
    
    """
    cur = conn.cursor()
    sql1 = ''' UPDATE stockprices SET date = ?,stockdetails_id = ?,openvalue=?,highvalue=?,
               lowvalue=?,closevalue=?,adjclose=?,volume=?,avg10volume=?,avg50volume=?,avg100volume=?,avg365volume=?, 
               minvolume=?,maxvolume=?,day50avgprice=?,day100avgprice=? WHERE date = ? and stockdetails_id =?''' 
    cur.execute(sql1, stockpricesinfo1)
        
    
def print_stockprices(conn):
    """
    Query all rows in the stockprices table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM stockprices")
    
    rows = cur.fetchall()
    print('\n'"stockprices Table")
    for row in rows:
        print(row)

def insert_stockvalueledger(conn,id,ownerid,portfolioid,currentvalue):
    """
    Insert a new stockvalueledger into the StockValueLedger table
    :param conn:
    :param stockid:
    :param registeredownerid:
    :param portfolioid:
    :param currentvalue(numberofstock * closevalueperstock):
    """
    cur = conn.cursor()
    count = cur.execute("SELECT count(*) FROM StockValueLedger WHERE stockdetails_id =?",
                        (id,))
    for row in count:
        
        if (row[0] == 0):
            sql = ''' INSERT INTO StockValueLedger(RegisteredOwnerID,PortfolioID,stockdetails_id,
                      PercentageOfOverallInvestmentByOwner,ProfitLossValueComparedtoCostInvestment,
                      ProfitLossPercentComparedtoCostInvestment,ProfitLossValueComparedtoPreviousClose,
                      ProfitLossPercentComparedtoPreviousClose,StockRiskCodeID,StockActionRecommendationCodeID,
                      CurrentValue,LastUpdatedDate,LastUpdatedTime)
                      VALUES(?,?,?,?,?,?,?,?,?,?,?,Date('now'),Time('now'))'''
            data =(ownerid,portfolioid,id,0,0,0,0,0,0,0,currentvalue) 
            cur.execute(sql,data) 
        else:
            sql = ''' UPDATE StockValueLedger SET CurrentValue=currentvalue,LastUpdatedDate=Date('now'),
                      LastUpdatedTime=Time('now')''' 
            cur.execute(sql)  

            
def print_stockvalueledger(conn):
    """
    Query all rows in the StockValueLedger table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM StockValueLedger")
    
    rows = cur.fetchall()
    print('\n'"StockValueLedger Table")
    for row in rows:
        print(row)
        
     
        
def create_connection(db):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    try:
        conn = sqlite3.connect(db)
        print("Database opened successfully")
        return conn
    except Error as e:
        print(e)
 
    return None

def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)
        
def close_connection(db_conn):
    """ Close connection to the database """
    db_conn.close()   
    
def insert_stockdetails(conn, stockinfo):
    """
    Create a new stockdetails into the stockdetails table
    :param conn:
    :param stockdetails:
    :return: id
    """
    sql = ''' INSERT INTO stockdetails(companysymbol,companyname)
              VALUES(?,?) '''
    
    cur = conn.cursor()
    cur.execute(sql, stockinfo)
    
    return cur.lastrowid 

                
def print_stockdetails(conn):
    """
    Query all rows in the stockdetails table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    cur.execute("SELECT * FROM stockdetails")
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
                
               
def main():
    
       
    database = "/Users/Ravi/Downloads/stock_backend/stocksqlite1.db"
 
    stockdetails_table = """CREATE TABLE IF NOT EXISTS stockdetails (
                                        id integer PRIMARY KEY,
                                        companysymbol varchar(10) NOT NULL,
                                        companyname text NOT NULL
                                        
                                    );"""                            
                                    
 
    stockprices_table = """CREATE TABLE IF NOT EXISTS stockprices (
                                    id integer PRIMARY KEY,
                                    date date NOT NULL,
                                    stockdetails_id integer NOT NULL,
                                    openvalue real NOT NULL,
                                    highvalue real NOT NULL,
                                    lowvalue real NOT NULL,
                                    closevalue real NOT NULL,
                                    adjclose real NOT NULL,
                                    volume BIGINT,
                                    avg10volume real, avg50volume real, avg100volume real, 
                                    avg365volume real,
                                    minvolume integer, maxvolume integer,
                                    day50avgprice real, day100avgprice real,
                                    FOREIGN KEY (stockdetails_id) REFERENCES stockdetails (id)
                                );"""
   
    registeredowners_table = """CREATE TABLE IF NOT EXISTS registeredowners (
                                    registeredownerid integer PRIMARY KEY,
                                    firebaseID varchar(50) NOT NULL,
                                    registeredowneremail text NOT NULL,
                                    newuser integer NOT NULL             
                                  );""" 
    
    PortfolioInventory_table = """ CREATE TABLE IF NOT EXISTS PortfolioInventory (
                                     PortfolioID integer PRIMARY KEY, 
                                     PortfolioName text NOT NULL,
                                     RegisteredOwnerID integer NOT NULL,
                                     PortfolioDescription text,
                                     TotalInvestedCost real,
                                     CurrentValue real,
                                     LastUpdatedDate date,
                                     LastUpdatedTime text,
                                     FOREIGN KEY (RegisteredOwnerID) REFERENCES registeredowners (registeredownerid)      
                                   ); """
    stockinventory_table = """ CREATE TABLE IF NOT EXISTS stockinventory (
                                   StockInvetoryID integer PRIMARY KEY,
                                   RegisteredOwnerID integer NOT NULL,
                                   PortfolioID integer NOT NULL,
                                   stockdetails_id integer NOT NULL,
                                   NumberOfStock real,
                                   AverageCost real,
                                   TotalInvestment real,
                                   CAGR real,
                                   Geomean real,
                                   PortfolioTaxTerm integer,
                                   ListOfTransaction text,
                                   LastUpdatedDate date,
                                   LastUpdatedTime text,
                                   FOREIGN KEY (stockdetails_id) REFERENCES stockdetails (id),
                                   FOREIGN KEY (RegisteredOwnerID) REFERENCES registeredowners (registeredownerid),
                                   FOREIGN KEY (PortfolioID) REFERENCES PortfolioInventory (PortfolioID)
                                  );"""
    
    StockValueLedger_table = """ CREATE TABLE IF NOT EXISTS StockValueLedger (
                                  StockValueLedgerID integer PRIMARY KEY,
                                  RegisteredOwnerID integer NOT NULL,
                                  PortfolioID integer NOT NULL, 
                                  stockdetails_id integer NOT NULL,
                                  PercentageOfOverallInvestmentByOwner integer,
                                  ProfitLossValueComparedtoCostInvestment integer,
                                  ProfitLossPercentComparedtoCostInvestment integer,
                                  ProfitLossValueComparedtoPreviousClose integer,
                                  ProfitLossPercentComparedtoPreviousClose integer,
                                  StockRiskCodeID integer,
                                  StockActionRecommendationCodeID integer,
                                  CurrentValue real,
                                  LastUpdatedDate date,
                                  LastUpdatedTime text,
                                  FOREIGN KEY (stockdetails_id) REFERENCES stockdetails (id),
                                  FOREIGN KEY (RegisteredOwnerID) REFERENCES registeredowners (registeredownerid),
                                  FOREIGN KEY (PortfolioID) REFERENCES PortfolioInventory (PortfolioID)
                                );"""
    # create a database connection
    conn = create_connection(database)
    if conn is not None:
        # create stockdetails table
        create_table(conn, stockdetails_table)
        
        # create stockprices table
        create_table(conn, stockprices_table)
        
        # create registeredowners table
        create_table(conn, registeredowners_table)
        
        # create stockinventory table
        create_table(conn, PortfolioInventory_table)
        
        
        # create stockinventory table
        create_table(conn, stockinventory_table)
        
        # create stockinventory table
        create_table(conn, StockValueLedger_table)
        
        
    else:
        print("Error! cannot create the database connection.")

    with conn:
        # create a new stockdetails
        
        #stockinfo = ('NVDA', 'Nvidia Corporataion');
        #id = insert_stockdetails(conn, stockinfo)
        print('\n'"stockdetails Table")
        print_stockdetails(conn)
        
        #while loop - get symbols from st
        # create a new stockprices
        #sym = "NVDA"
        
        cur = conn.cursor()
        cur.execute("SELECT stockdetails_id FROM stockinventory")
        rows = cur.fetchall()
        
        for row in rows:
            id = row[0]
            sym = get_symbol(conn, id)
            stockprices = update_data_to_db(conn, sym)
        print_stockprices(conn)
        
        totalvalue = 0;
        #Create a new StockValueLedger
        cur = conn.cursor()
        cur.execute("SELECT stockdetails_id,NumberOfStock,stockinventory.RegisteredOwnerID,"
                    "stockinventory.PortfolioID FROM stockinventory INNER JOIN PortfolioInventory "
                     "ON PortfolioInventory.PortfolioID = stockinventory.PortfolioID")
        rows = cur.fetchall()
        
        for row in rows:
            id = row[0]
            sym = get_symbol(conn, id)
            closevalue = update_currentvalue(conn, sym)
            currentvalue = row[1] * closevalue
            totalvalue = totalvalue + currentvalue
            insert_stockvalueledger(conn,id,row[2],row[3],currentvalue)
            print("symbol&currentvalue",sym,currentvalue)
        print("totalvalue",totalvalue)    
        print_stockvalueledger(conn)
        
    # close database connection      
    close_connection(conn)

if __name__ == '__main__':
    main()
```
